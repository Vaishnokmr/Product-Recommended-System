```python
stat_book.describe()
```

```python
y = stat_book[stat_book>35]
limelight_books = y.index
```

```python
final = fitered[fitered["Book-Title"].isin(limelight_books)]
```

```python
pivot_final = final.pivot_table(index='Book-Title', columns='User-ID',values='Book-Rating')
```

```python
pivot_final.fillna(0, inplace=True)
```

```python
pivot_final
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pivot_final)
```

```python
similarity_scores.shape
```

### function

```python
def recommend(book_name):
    index = np.where(pivot_final.index == book_name)[0][0]
    similar_item = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    data = []
    for i in similar_item:
        temp_df =  books[books['Book-Title'] == pivot_final.index[i[0]]]
        item = []
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
    
    
```

```python
recommend("The Handmaid's Tale")
```
